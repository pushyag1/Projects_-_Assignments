```python
##total count of the Years

dfYear= pd.value_counts(data1['Year'].values)
dfYear
```

*segregation of data:*

sum1 represents the sum of 'HIV related death rates' for the respective years.

sum2 represents the sum of 'Non-HIV related death rates' for the respective years.

```python
sum1=data1.groupby('Year', as_index=False).agg({"HIV-related death rate": "sum"})
sum1

#citation: https://www.geeksforgeeks.org/python-pandas-dataframe-groupby/
#citation: https://stackoverflow.com/questions/36684013/extract-column-value-based-on-another-column-pandas-dataframe
#citation :https://www.geeksforgeeks.org/python-pandas-dataframe-add/
#citation: https://www.shanelynn.ie/summarising-aggregation-and-grouping-data-in-python-pandas/
```

```python
sum2=data1.groupby('Year', as_index=False).agg({"Non-HIV-related death rate": "sum"})
sum2
#citation: https://www.geeksforgeeks.org/python-pandas-dataframe-groupby/
#citation: https://stackoverflow.com/questions/36684013/extract-column-value-based-on-another-column-pandas-dataframe
#citation :https://www.geeksforgeeks.org/python-pandas-dataframe-add/
#citation: https://www.shanelynn.ie/summarising-aggregation-and-grouping-data-in-python-pandas/
```

```python

sum3=data1.groupby('Year', as_index=False).agg({"Death rate": "sum"})
sum3
#citation: https://www.geeksforgeeks.org/python-pandas-dataframe-groupby/
#citation: https://stackoverflow.com/questions/36684013/extract-column-value-based-on-another-column-pandas-dataframe
#citation :https://www.geeksforgeeks.org/python-pandas-dataframe-add/
#citation: https://www.shanelynn.ie/summarising-aggregation-and-grouping-data-in-python-pandas/
```

```python
#checking if the data got aligned properly.

df=[sum1,sum2,sum3]
#df
result = pd.concat(df)
result

#citation :https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html
#citation : https://datacarpentry.org/python-ecology-lesson/05-merging-data/
```

```python
#merging two data frames using inner join and assigning it to df1

df1 = pd.merge(sum1, sum2, on='Year',how='inner')
df1

#citation: https://stackoverflow.com/questions/41815079/pandas-merge-join-two-data-frames-on-multiple-columns
```

```python
#merging two data frames (df1 and sum3) using inner join and assigning it to data2

data2 = pd.merge(df1, sum3, on='Year',how='inner')
data2

#citation: https://stackoverflow.com/questions/41815079/pandas-merge-join-two-data-frames-on-multiple-columns
```
