### Question 1:

> * Which category of people are prone to mental health and also the suicide victims in Maryland?

```python
#df3 represents sum of all the values for each Race/ ethnicity for Mental Health ED dataset
#renaming the columns
df3 = d1_notstate.groupby("Race/ ethnicity" ,as_index = False).agg({"Value": "sum"})
df3.columns = ["Race/ ethnicity","Mental Health ED"]
df3
```

```python
df4 = d2_notstate.groupby("Race/ ethnicity" ,as_index = False).agg({"Value": "sum"})
df4.columns = ["Race/ ethnicity","suicide rate"]
df4
```

```python
r_values = df3['Mental Health ED'][0],df3['Mental Health ED'][1:3].sum(), df3['Mental Health ED'][3:5].sum(),df3['Mental Health ED'][5],df3['Mental Health ED'][6:8].sum()
re_values_m = pd.DataFrame(r_values)
re_values_m.columns = ['Mental Health ED']
re_values_m
r_e = ['others (aggregated)','Asian/ Pacific Islander(aggregated)','Black (aggregated)','Hispanic','White (aggregated)']
re_values_m ['Race/ ethnicity'] = r_e
re_values_m
```

```python
s_values = df4['suicide rate'][0],df4['suicide rate'][1],df4['suicide rate'][2],df4['suicide rate'][3], df4['suicide rate'][4]
s_values
re_values_s = pd.DataFrame(s_values)
re_values_s.columns = ['suicide rate']
re_values_s
s_e = ['others (aggregated)','Asian/ Pacific Islander(aggregated)','Black (aggregated)','Hispanic','White (aggregated)']
re_values_s ['Race/ ethnicity'] = s_e
re_values_s
```

```python
df_ethnicity = pd.merge(re_values_m,re_values_s,on ="Race/ ethnicity", how="inner")
df_ethnicity
```

```python
#converting the suicide rate values and Mental Health ED values to percentages for easy comparision

df_ethnicity["suicide rate"] = df_ethnicity["suicide rate"].div(df_ethnicity["suicide rate"].sum()).multiply(100)
df_ethnicity["Mental Health ED"] = df_ethnicity["Mental Health ED"].div(df_ethnicity["Mental Health ED"].sum()).multiply(100)
df_ethnicity
```

```python
data_data=df_ethnicity.set_index('Race/ ethnicity').T
```

```python
#Plotting a percentage stacked graph

colours = plt.cm.PiYG(np.linspace(0, 1, 10))
ax = data_data.plot(kind='bar', stacked= True, figsize=(20,10) , color = colours)
ax.grid(False)
ax.set_title('Race/ ethnicity prone to mental health and also suicide victims in Maryland jurisdictions',fontsize ='30')
ax.set_xlabel ('mental and suicide rates', fontsize ='20')
ax.set_ylabel ('percentage', fontsize ='20')
plt.legend(loc = "best")

#https://pstblog.com/2016/10/04/stacked-charts
```

```python
#stacked graph for individual races.

f, ax = plt.subplots(1, figsize=(20,10))
bar_width = 1
bar_l = [i for i in range(len(df_ethnicity['Mental Health ED']))]
tick_pos = [i+(bar_width/2) for i in bar_l] 
totals = [i+j for i,j in zip(df_ethnicity['Mental Health ED'], df_ethnicity['suicide rate'])]
pre_rel = [i / j * 100 for  i,j in zip(df_ethnicity['Mental Health ED'], totals)]
mid_rel = [i / j * 100 for  i,j in zip(df_ethnicity['suicide rate'], totals)]
ax.bar(bar_l, pre_rel, label='Mental Health ED', alpha=0.9,color='skyblue',edgecolor='black')
ax.bar(bar_l,mid_rel, bottom=pre_rel, label='suicide rate',alpha=0.9,color='#7FFFD4',hatch = '/',edgecolor='black')
plt.xticks(tick_pos, df_ethnicity['Race/ ethnicity'])
#ax.grid(False)
ax.set_title('Comparing which Race/Ethnicity has got go more suide and mental values individually' ,fontsize =25)
ax.set_ylabel("Percentage")
ax.set_xlabel("")
plt.xlim([min(tick_pos)-bar_width, max(tick_pos)+bar_width])
plt.ylim(-10, 110)
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
plt.legend(loc = "upper right")
plt.show()

#citation: https://pstblog.com/2016/10/04/stacked-charts
```

According to the visualizations, Blacks are having more mental health rates where as Asian/pacific Islanders are having high suicide rates when compared to other races/ethnicities.

### Question 3:

>* Do health condition correspond impact on the Suicide Rates in Maryland?

```python
df1 = d1_notstate.groupby("Jurisdiction" ,as_index = False).agg({"Value": "sum"})
df1.columns=["Jurisdiction", "Mental Health ED"]
df1
#df2 represents sum of all the values for each Jurisdiction for suicide rate dataset

df2 = d2_notstate.groupby("Jurisdiction", as_index= False).agg({"Value":"sum"})
df2.columns=["Jurisdiction", "suicide rate value"]
df2
#merging df1 and df2 using inner join on jurisdiction and assigning to df_value

df_value = pd.merge(df1, df2, on = "Jurisdiction", how ="inner")
df_value
```

```python
#The values of Mental Health ED values are very high when compared to suicide rate value, so converted the values into percentages
#to properly compare them

df_value["Mental Health ED"] = df_value["Mental Health ED"].div(df_value["Mental Health ED"].sum()).multiply(100)
df_value["suicide rate value"] = df_value["suicide rate value"].div(df_value["suicide rate value"].sum()).multiply(100)

#citation: https://stackoverflow.com/questions/42006346/pandas-convert-columns-to-percentages-of-the-totals/42006745
```

```python
#checking for highly effected Jurisdiction for suicide rate also have the highly effected value for mental health ED
fig, ax = plt.subplots(figsize=(35,8))
plt.plot('Jurisdiction' ,'Mental Health ED', data=df_value, marker='', markerfacecolor='blue', color='skyblue', linewidth=4)
plt.plot( 'Jurisdiction','suicide rate value',  data=df_value, marker='', color='red', linewidth=3)
plt.xlabel('Jurisdiction', fontsize=23)
plt.ylabel('values', fontsize=23)
plt.title('Increase or Decrease (behavioural of suicide and mental health rates of all the jurisdictions)', fontsize=29)
plt.legend()

#citation: https://python-graph-gallery.com/line-chart/
#citation: http://www.learningaboutelectronics.com/Articles/How-to-change-the-line-width-of-a-graph-plot-in-matplotlib-with-Python.php
```
