Simulate your course grade in a Jupyter notebook using Python3.

Suppose the course grade is based on 24 assignments (each scored 0 to 100), 2 quizzes (each scored 0 to 40), three projects (each scored 0 to 100), and attendance in 15 classes (each scored as present or absent). The assignments+quizzes account for 30% of the course grade, the projects are 20% each (60% total), and attendance is 10% of the course grade. Suppose 10 of the (24 assignments+2 quizzes) are included in the final course score. The 10 are randomly selected. Letter grades are defined in the syllabusâ€‹.

Use your (known) project scores, attendance, quiz scores, and assignment scores. Use your guess for your scores for the remaining homeworks and project. Simulate selecting randomly 10 assignment scores to get your course grade (in percentage points). Repeat the simulation 100 times (each with a different random selection of assignments). Submit a notebook that performs this simulation and address the following objectives:

Plot the distribution of your course grade for the 100 simulated outcomes.
What is the maximum course grade, minimum course grade, and average course grade?

```python
#importing necessary libraries

import random
import pandas as pd
import matplotlib.pyplot as plt
from statistics import mean
import collections

#citation : https://www.geeksforgeeks.org/python-statistics-mean-function/
```

```python
quiz1 = 30
q1 = (30 /40) * 100
print(q1)
quiz2 = 40
q2 = (40/40) *100
print(q2)
```

 > * calculating attendance percentage for 15 classes

```python
att = ['present','present','present','present','present','present','present','present','present','present','present','present','present','present','present',]
attend= att.count('present')
print('number of classes attended out of 15:' , attend)
attendance = (((attend/15)*100)*0.1)
print("attendance %:",attendance)
```

### for 100 distributions

> * selecting 10 random assignments including quiz
> * adding them and taking the mean
> * taking all the 3 projects
> * attendance = 10% (100% attendance)
> * calculating the overall percentage 
> * calculating the grade for each overall percent

```python
#quizes : scored 40 and 30, coverted to 100 and added in assignments
#24 assignments + 2 quizes = 26 for random choices
n = 100
t =[]
grade = []
c= []
for x in range(n):
    Assignments  = [100,75,85,100,96,100,100,85,100,100,85,100,100,80,100,90,100,85,q1,q2,100,100,100,95,100,100]
    choice = random.sample(Assignments, k=10)
    print("random 10 assignments ", choice)
    random_Assigns = sum(choice)
    c.append(choice)
    print ("sum_of randomly picked assignments = ",random_Assigns)
    assign_percent = float(((random_Assigns/1000 )*100)*0.3)
    a_percent = round(assign_percent , 2)
    print("average of randomly picked assignments = ",a_percent)
    project1 = 99      
    project2 = 100     
    project3 = 70
    print("project marks:","project1 : ",project1,"project 2 :", project2, "project 3 :",project3)
    project1_Percent = round(float(((project1 / 100) * 0.2)*100), 2)
    project2_Percent = round(float(((project2 / 100) * 0.2)*100), 2)
    project3_Percent = round(float(((project3 / 100) * 0.2)*100), 2)
    #print(project1_Percent, project2_Percent, project3_Percent)
    print("project percentage:","project1 : ",project1_Percent,"project 2 :", project2_Percent, "project 3 :",project3_Percent)
    projects = project1_Percent + project2_Percent + project3_Percent
    print("over all project percentage for 60 :" ,projects )
    #attendance calculated above
    print("attendance %:",attendance)
    s = round((projects + a_percent + attendance) , 2)
    print("overall percentage : ",s)
    t.append(s)
    if s>=90:
        Grade = 'A'
    elif s>=80:
        Grade = 'B'
    elif s>=70:
        Grade = 'C'
    elif s>=60:
        Grade = 'D'
    else:
        Grade ="F"
    grade.append(Grade)
    print("your grade is : ", Grade)
    
# citation: https://www.programiz.com/python-programming/if-elif-else
#citation : https://pynative.com/python-random-choice/
# citation : https://stackoverflow.com/questions/6494508/how-do-you-pick-x-number-of-unique-numbers-from-a-list-in-python   
```
