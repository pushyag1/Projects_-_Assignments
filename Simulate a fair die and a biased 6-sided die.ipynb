```python
#importing random, numpy, pandas, matplotlib, seaborn

import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#citations: https://jakevdp.github.io/PythonDataScienceHandbook/04.05-histograms-and-binnings.html
#citations: https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html
```

```python
#asking the user for number of rolls and taking fair dice rolls and biased dice rolls

df = []
List = ['bd', 'fd']
number=int(input("number of rolls ="))    
bias = [0.15, 0.15, 0.15, 0.15, 0.15, 0.25]   # a 6-sided biased die
np.where(np.random.multinomial(1, bias, size=1)[0]==1)[0][0]+1    
for i in range(number): 
        fd=[random.randint(1,6) for i in range(number) ]
def roll(probs, ntimes):   # roll a len(probs) sided biased die with bias probs for ntimes
    return np.apply_along_axis(lambda x: x.tolist().index(1)+1, 1, 
                       np.random.multinomial(1, bias,  size=number)).tolist()

bd=roll(probs=bias, ntimes=number)   

#citation:https://stackoverflow.com/questions/17368554/python-how-to-simulate-the-roll-of-a-dice-100-times
#citation:https://stackoverflow.com/questions/479236/how-do-i-simulate-biased-die-in-python
#Citation:https://stackoverflow.com/questions/44008489/dice-rolling-simulator-in-python
```
